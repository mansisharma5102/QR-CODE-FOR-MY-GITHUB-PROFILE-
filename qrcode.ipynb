```python
#importing libraries
import qrcode
from PIL import Image

# changing functionality of the code

qr = qrcode.QRCode(
    version =1,
    error_correction= qrcode.constants.ERROR_CORRECT_H,
    box_size=10,
    border=4

)

#Link of website for which you are creating qr code

qr.add_data("https://github.com/mansisharma5102")
qr.make(fit =True)
img = qr.make_image(fill_color = "#000080", back_color = "white")
img.save("mansisharma.png")
img.show()
```
